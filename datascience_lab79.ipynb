```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
```

```python
#histogram
plt.figure(figsize=(6,4))
sns.histplot(df['sepal length (cm)'], bins=20, kde=True)
plt.title("Histogram of Sepal Length")
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(6,4))
sns.boxplot(x=df['sepal length (cm)'])
plt.title("Boxplot of Sepal Length")
plt.show()
```

```python
# Pie Chart (species distribution)
species_counts = df['species'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(species_counts, labels=iris.target_names, autopct='%1.1f%%')
plt.title("Species Distribution (Pie Chart)")
plt.show()
```

bivariate visualizations

```python
# Scatterplot
plt.figure(figsize=(6,4))
sns.scatterplot(x=df['sepal length (cm)'], y=df['sepal width (cm)'], hue=df['species'])
plt.title("Scatterplot: Sepal Length vs Sepal Width")
plt.show()
```

```python
# line Chart (trend of sepal length across samples)
plt.figure(figsize=(6,4))
plt.plot(df['sepal length (cm)'])
plt.title("Line Chart of Sepal Length")
plt.xlabel("Sample Index")
plt.ylabel("Sepal Length (cm)")
plt.show()
```

```python
# Bar Chart (mean sepal length per species)
plt.figure(figsize=(6,4))
sns.barplot(x=df['species'], y=df['sepal length (cm)'])
plt.title("Bar Chart: Mean Sepal Length by Species")
plt.xticks([0,1,2], iris.target_names)
plt.show()
```

multivariate visualizations


```python
# Heatmap (correlation matrix)
plt.figure(figsize=(8,6))
sns.heatmap(df.iloc[:,:4].corr(), annot=True, cmap="coolwarm")
plt.title("Heatmap of Feature Correlations")
plt.show()
```

```python
# Bubble Chart (scatterplot with size = petal length)
plt.figure(figsize=(6,4))
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'],
            s=df['petal length (cm)']*20, alpha=0.5, c=df['species'])
plt.title("Bubble Chart: Sepal vs Petal (size = petal length)")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.show()
# Pair Plot (relationships among all features)
sns.pairplot(df.iloc[:,:4])
plt.suptitle("Pair Plot of Iris Features", y=1.02)
plt.show()
```

```python
# Violin Plot
plt.figure(figsize=(6,4))
sns.violinplot(x=df['species'], y=df['sepal length (cm)'])
plt.title("Violin Plot: Sepal Length by Species")
plt.xticks([0,1,2], iris.target_names)
plt.show()
```

```python
#bernoulli distribution
import numpy as np
from scipy.stats import bernoulli, binom

print("----- Bernoulli Distribution -----")
p = 0.7

bernoulli_samples = bernoulli.rvs(p, size=10)
print("Probability of Passing:", p)
print("Student Outcomes (1=Pass, 0=Fail):")
print(bernoulli_samples)

print("\nProbability of Pass:")
print(bernoulli.pmf(1, p))
print("Probability of Fail:")
print(bernoulli.pmf(0, p))


```

```python
#binomial distribution
print("\n----- Binomial Distribution -----")
n = 10
k = 8

binomial_probability = binom.pmf(k, n, p)
print("Number of Students:", n)
print("Probability of Passing:", p)
print(f"Probability that exactly {k} students pass:")
print(binomial_probability)

random_binomial = binom.rvs(n, p, size=5)
print("Random Binomial Outcomes:")
print(random_binomial)
```

```python
#Common Discrete Distributions
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom, geom, poisson

# Bernoulli
p = 0.6
x = np.array([0,1])
plt.bar(x, bernoulli.pmf(x,p))
plt.title("Bernoulli")
plt.show()

# Binomial
n = 10
p = 0.5
x = np.arange(0,n+1)
plt.bar(x, binom.pmf(x,n,p))
plt.title("Binomial")
plt.show()

# Geometric
p = 0.3
x = np.arange(1,11)
plt.bar(x, geom.pmf(x,p))
plt.title("Geometric")
plt.show()

# Poisson
lam = 4
x = np.arange(0,15)
plt.bar(x, poisson.pmf(x,lam))
plt.title("Poisson")
plt.show()
```

```python
#Common Continuous Distributions
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, uniform, expon, chi2, t

# Normal
x = np.linspace(-4,4,200)
plt.plot(x, norm.pdf(x))
plt.title("Normal")
plt.show()

# Uniform
x = np.linspace(0,10,200)
plt.plot(x, uniform.pdf(x,0,10))
plt.title("Uniform")
plt.show()

# Exponential
x = np.linspace(0,10,200)
plt.plot(x, expon.pdf(x))
plt.title("Exponential")
plt.show()

# Chi-square
x = np.linspace(0,20,200)
plt.plot(x, chi2.pdf(x,5))
plt.title("Chi-Square")
plt.show()

# t-distribution
x = np.linspace(-5,5,200)
plt.plot(x, t.pdf(x,10))
plt.title("t Distribution")
plt.show()
```

```python
#Z-Test
import numpy as np
from statsmodels.stats.weightstats import ztest

sample_marks = np.array([78,82,85,88,90,76,95,89,84,91])
population_mean = 80

z_stat, p_value = ztest(sample_marks, value=population_mean)

print("Z-Statistic:", z_stat)
print("P-Value:", p_value)

if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

```python
#Independent t-Test
import numpy as np
from scipy import stats

class_A = np.array([78,82,85,88,90])
class_B = np.array([72,75,70,68,74])

t_stat, p_value = stats.ttest_ind(class_A, class_B)

print("t-Statistic:", t_stat)
print("P-Value:", p_value)

if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

```python
#Parametric Tests
import numpy as np
from scipy import stats

group1 = np.random.normal(70,10,30)
group2 = np.random.normal(75,10,30)
group3 = np.random.normal(80,10,30)

print(stats.ttest_ind(group1, group2))
print(stats.f_oneway(group1, group2, group3))
```

```python
#Non-Parametric Tests
import numpy as np
from scipy import stats

group1 = np.array([45,50,52,55,58])
group2 = np.array([48,53,57,59,61])

print(stats.mannwhitneyu(group1, group2))

before = np.array([50,52,55,57])
after = np.array([53,55,56,60])

print(stats.wilcoxon(before, after))
```
