```python
#mean median mode
import pandas as pd
import numpy as np
df = pd.DataFrame({  'Age': [25, 30, np.nan, 40, 35],
    'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']})
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Department'].fillna(df['Department'].mode()[0], inplace=True)
print(df)
```

```python
#forward fill
import pandas as pd
import numpy as np
df = pd.DataFrame({  'Age': [25, 30, np.nan, 40, 35],
'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']})
df_ffill = df.copy()
df_ffill.fillna(method='ffill', inplace=True)
print(df_ffill)
```

```python
#backward fill
import pandas as pd
import numpy as np
df = pd.DataFrame({  'Age': [25, 30, np.nan, 40, 35],
'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']})
df_bfill = df.copy()
df_bfill.fillna(method='bfill', inplace=True)
print(df_bfill)
```

```python
#deletion(drop rows)
import pandas as pd
import numpy as np
df = pd.DataFrame({  'Age': [25, 30, np.nan, 40, 35],
'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']})
df_drop_rows=df.dropna()
print("After dropping rows:")
print(df_drop_rows)
```

```python
#deletion(drop columns)
import pandas as pd
import numpy as np
df = pd.DataFrame({  'Age': [25, 30, np.nan, 40, 35],
'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']})
df_drop_columns=df.dropna(axis=1)
print("After dropping columns:")
print(df_drop_columns)
```

```python
#drop rows that are identical across all columns
import pandas as pd
import numpy as np
df = pd.DataFrame({
    'ID': [1, 2, 2, 3, 4, 4],
    'Name': ['Alice', 'Bob', 'Bob', 'Charlie', 'David', 'David'],
    'Age': [25, 30, 30, 35, 40, 40]
})

print("Original DataFrame:")
print(df)

df_exact = df.drop_duplicates()
print("\nAfter Exact Match Removal:")
print(df_exact)
```

```python
#data standardization

import pandas as pd
df = pd.DataFrame({
    'Date': ['2025-01-05', '05/01/2025', 'Jan 5, 2025', '2025.01.05']
})

df['Date'] = pd.to_datetime(df['Date'], errors='coerce').dt.strftime('%Y-%m-%d')

print(df)
```

```python
#case normalisation
df=pd.DataFrame({
    'Name': ['Alice', 'BOB', 'charlie', 'DAVID']
})

df['Name_lower'] = df['Name'].str.lower()

df['Name_upper'] = df['Name'].str.upper()

print(df)
```
