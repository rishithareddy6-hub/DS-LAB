```python
#regression models for predictive analysis
import numpy as np
from sklearn.linear_model import LinearRegression, LogisticRegression

print("----- Linear Regression -----")
X_linear = np.array([[1000], [1200], [1500], [1800], [2000]])
y_linear = np.array([50, 55, 65, 80, 90])
linear_model = LinearRegression()
linear_model.fit(X_linear, y_linear)
new_house_size = np.array([[1700]])
predicted_price = linear_model.predict(new_house_size)
print("Training House Sizes:")
print(X_linear.flatten())
print("\nTraining House Prices:")
print(y_linear)
print("\nPredicted Price for 1700 sq.ft house:")
print(predicted_price[0])
print("\n\n----- Logistic Regression -----")
X_logistic = np.array([[40], [50], [60], [70], [80], [90]])

y_logistic = np.array([0, 0, 0, 1, 1, 1])
logistic_model = LogisticRegression()
logistic_model.fit(X_logistic, y_logistic)
test_price = np.array([[75]])
purchase_prediction = logistic_model.predict(test_price)
purchase_probability = logistic_model.predict_proba(test_price)
print("Training House Prices:")
print(X_logistic.flatten())
print("\nPurchase Labels:")
print(y_logistic)
print("\nTest House Price:", test_price[0][0])
print("\nPurchase Prediction:")
print(purchase_prediction[0])
print("\nPrediction Probabilities:")
print(purchase_probability)
```

```python
#classification using decision tree
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix
)

X = np.array([
    [2, 50],
    [3, 60],
    [4, 65],
    [5, 70],
    [6, 75],
    [7, 80],
    [8, 85],
    [9, 90]
])
y = np.array([0, 0, 0, 1, 1, 1, 1, 1])
model = DecisionTreeClassifier(random_state=42)

model.fit(X, y)
X_test = np.array([
    [3, 55],
    [5, 72],
    [8, 88],
    [4, 60]
])
y_test = np.array([0, 1, 1, 0])
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
error_rate = 1 - accuracy
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print("Actual Labels:")
print(y_test)
print("\nPredicted Labels:")
print(y_pred)
print("\nConfusion Matrix:")
print(cm)
print("\nAccuracy:")
print(accuracy)
print("\nError Rate:")
print(error_rate)
print("\nPrecision:")
print(precision)
print("\nRecall:")
print(recall)
print("\nF1-Score:")
print(f1)
```

```python
#knn classfication
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix
)
X_train = np.array([
    [2, 50],
    [3, 60],
    [4, 65],
    [5, 70],
    [6, 75],
    [7, 80],
    [8, 85],
    [9, 90]
])
y_train = np.array([0, 0, 0, 1, 1, 1, 1, 1])
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
X_test = np.array([
    [3, 55],
    [5, 72],
    [8, 88],
    [4, 60]
])
y_test = np.array([0, 1, 1, 0])
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
error_rate = 1 - accuracy
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
print("Actual Labels:")
print(y_test)
print("\nPredicted Labels:")
print(y_pred)
print("\nConfusion Matrix:")
print(cm)
print("\nAccuracy:")
print(accuracy)
print("\nError Rate:")
print(error_rate)
print("\nPrecision:")
print(precision)
print("\nRecall:")
print(recall)
print("\nF1-Score:")
print(f1)
```
